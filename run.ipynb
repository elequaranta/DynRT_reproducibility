# Run

Get the last checkpoint and set the config, so that the algorithm will continue training from it.

```python
import os
import json

run_folders = [x for x in os.listdir("exp")]

checkpoint = None

for i in range(len(run_folders)):
  curr_run_folder = run_folders[len(run_folders) - 1 - i]
  if 'model_best.pth.tar' in os.listdir("exp/" + curr_run_folder + "/checkpoints"):
    checkpoint = "exp/" + curr_run_folder + "/checkpoints/model_best.pth.tar"
    break

print(checkpoint)

with open('config/DynRT.json', 'r+') as f:
    data = json.load(f)
    if checkpoint is None:
      data['info']['train_on_checkpoint'] = "none"
    else:
      data['info']['train_on_checkpoint'] = checkpoint
    f.seek(0)
    json.dump(data, f, indent=4)
    f.truncate()
```
