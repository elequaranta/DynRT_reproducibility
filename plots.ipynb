```python
import matplotlib.pyplot as plt

```

```python
def plot_loss(loss_values, title='Loss Plot'):
    plt.plot(loss_values, label='Loss')
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

```python

loss_values = [None, None, 0.616, 0.6105, 0.6065, 0.6031, 0.5994, 0.5960, 0.5896, 0.5793, 0.5553, 0.4177, 0.3658, 0.3629, 0.3618, 0.3617]

plot_loss(loss_values, title='Loss Over Epochs')
```

```python
def plot_metrics(train_values, test_values, title, metric_name):
    plt.plot(train_values, label='Train')
    plt.plot(test_values, label='Test')
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel(metric_name)
    plt.legend()
    plt.show()
```

```python
train_accuracy = [None, None, 0.6797, 0.6942, 0.6993, 0.705, 0.7079, 0.7133, 0.7207, 0.7304, 0.7438, 0.8062, 0.9642, 0.9821, 0.9829, 0.9828]
test_accuracy = [None, None, 0.689, 0.6865, 0.6924, 0.6979, 0.705, 0.7101, 0.7181, 0.7269, 0.7547, 0.9111, 0.9347, 0.9355, 0.9359, 0.9359]
plot_metrics(train_accuracy, test_accuracy, 'Accuracy Over Epochs', 'Accuracy')
```

```python
plt.plot(train_f1, label='Train F1', color='blue')
plt.plot(test_f1, label='TestF1', color='green')
plt.plot(train_accuracy, label='Train Accuracy', color='red')
plt.plot(test_accuracy, label='Test Accuracy', color='orange')
plt.title('Accuracy and F1 Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy, F1')
plt.legend()
plt.show()
```
